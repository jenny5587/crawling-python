```python
import time
import re
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import csv
import pandas as pd

def wait_elem(browser, x_path, delay=10):
    wait_for = EC.presence_of_element_located((By.XPATH, x_path))
    ret = WebDriverWait(browser, delay).until(wait_for)
    return ret

def get_elem(browser, x_path):
    return browser.find_element(By.XPATH, x_path)

service = Service(ChromeDriverManager().install())
browser = webdriver.Chrome(service=service)

browser.execute_cdp_cmd("Page.addScriptToEvaluateOnNewDocument", {
    "source": """
            Object.defineProperty(navigator, 'webdriver', {
                get: () => undefined
            })
            """
})

# Open the Coupang product review page
browser.get("https://www.coupang.com/vp/products/6840775558?itemId=16266042968&vendorItemId=83458832307&q=%EB%A6%B4&itemsCount=36&searchId=b99c546f19c7480e8aab2518e57fa86d&rank=1&isAddedCart=#sdpReview")

# Perform login if necessary
id_input = get_elem(browser, '//*[@id="login-email-input"]')
c_id = ""
id_input.send_keys(c_id)

pw_input = get_elem(browser, '//*[@id="login-password-input"]')
c_pw = ""
pw_input.send_keys(c_pw)

get_elem(browser, '/html/body/div[1]/div/div[1]/form/div[6]/button').click()

time.sleep(3)

# Switch to the review tab
element = get_elem(browser, '//*[@id="btfTab"]/ul[1]/li[2]')
browser.execute_script("arguments[0].click();", element)

time.sleep(5)

html = browser.page_source

soup = BeautifulSoup(html, "html.parser")

elements= soup.find_all(class_="sdp-review__article__list js_reviewArticleReviewList")

# Initialize an empty list to store the extracted data
data_list = []

time.sleep(3)
# Iterate over the elements and extract specific data
for element in elements:
    # Extract specific data from each element
    title = element.find('div', class_ = 'class="sdp-review__article__list__headline')
    order_date = element.find('div', class_='sdp-review__article__list__info__product-info__reg-date').text
    review_contents = element.find('div', class_='sdp-review__article__list__review__content js_reviewArticleContent').text

    # Create a dictionary representing a row of data
    data_dict = {'title': title, 'Order Date': order_date, 'Review Contents': review_contents}

    # Append the dictionary to the data list
    data_list.append(data_dict)

    # Convert the list of dictionaries into a DataFrame
    df = pd.DataFrame(data_list)

    time.sleep(3)
        
# Print the resulting DataFrame
print(df)
```

```python
import time
import re
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import csv
import pandas as pd

def wait_elem(browser, x_path, delay=10):
    wait_for = EC.presence_of_element_located((By.XPATH, x_path))
    ret = WebDriverWait(browser, delay).until(wait_for)
    return ret

def get_elem(browser, x_path):
    return browser.find_element(By.XPATH, x_path)

service = Service(ChromeDriverManager().install())
browser = webdriver.Chrome(service=service)

browser.execute_cdp_cmd("Page.addScriptToEvaluateOnNewDocument", {
    "source": """
            Object.defineProperty(navigator, 'webdriver', {
                get: () => undefined
            })
            """
})

# Open the Coupang product review page
browser.get("https://www.coupang.com/vp/products/6840775558?itemId=16266042968&vendorItemId=83458832307&q=%EB%A6%B4&itemsCount=36&searchId=b99c546f19c7480e8aab2518e57fa86d&rank=1&isAddedCart=#sdpReview")

# Perform login if necessary
id_input = get_elem(browser, '//*[@id="login-email-input"]')
c_id = ""
id_input.send_keys(c_id)

pw_input = get_elem(browser, '//*[@id="login-password-input"]')
c_pw = ""
pw_input.send_keys(c_pw)

get_elem(browser, '/html/body/div[1]/div/div[1]/form/div[6]/button').click()

time.sleep(3)

# Switch to the review tab
element = get_elem(browser, '//*[@id="btfTab"]/ul[1]/li[2]')
browser.execute_script("arguments[0].click();", element)

time.sleep(5)

html = browser.page_source

soup = BeautifulSoup(html, "html.parser")


# Initialize an empty list to store the extracted data
data_list = []

time.sleep(5)

for k in range(2,23):

    print("페이지 수 : ",k)
    print("--------------------------------")
    # browser.find_element(By.XPATH,f'//*[@id="btfTab"]/ul[2]/li[3]/div/div[6]/section[4]/div[3]/button[%d]' %k).send_keys(Keys.ENTER)
#         browser.find_element(By.XPATH,f'//*[@id="btfTab"]/ul[2]/li[3]/div/div[6]/section[4]/div[3]/button[{k}]').send_keys(Keys.ENTER)
    if k >= 13:
        # //*[@id="btfTab"]/ul[2]/li[3]/div/div[6]/section[4]/div[3]/button[11]
        # //*[@id="btfTab"]/ul[2]/li[3]/div/div[6]/section[4]/div[3]/button[12]
        # //*[@id="btfTab"]/ul[2]/li[3]/div/div[6]/section[4]/div[3]/button[2]
        get_elem(browser, '//*[@id="btfTab"]/ul[2]/li[3]/div/div[6]/section[4]/div[3]/button[%d]' %(k-11)).click()

    else:
        print(k)
        get_elem(browser, '//*[@id="btfTab"]/ul[2]/li[3]/div/div[6]/section[4]/div[3]/button[%d]' %k).click()

    time.sleep(5)
    # data_list = []
    # Iterate over the elements and extract specific data
    with open("crawl.html", "w", encoding="utf-8") as f:
        f.write(browser.page_source)

    # HTML 파일 읽고 문자열 리턴
    with open('crawl.html', 'r', encoding='utf-8') as f:
        page = f.read()

    soup = BeautifulSoup(page, 'html.parser')  # Soup 객체 생성
    elements= soup.find_all(class_="sdp-review__article__list js_reviewArticleReviewList")
    time.sleep(3)

    for ele in elements:
        data_dict = {}
        # Extract specific data from each element
        order_date = ele.find('div', class_='sdp-review__article__list__info__product-info__reg-date').text
        header_contents = ele.find('div', class_='sdp-review__article__list__headline')
        # review_contents = ele.find('div', class_='sdp-review__article__list__review__content js_reviewArticleContent').text
        review_contents = ele.find('div', class_='sdp-review__article__list__review__content js_reviewArticleContent')
        if isinstance(header_contents, type(None)):
            header = ""
        else:
            header = ele.find('div', class_='sdp-review__article__list__headline').text
        if isinstance(review_contents, type(None)):
            review = ""

        else:
            review = ele.find('div', class_='sdp-review__article__list__review__content js_reviewArticleContent').text

        # Create a dictionary representing a row of data
        data_dict = {'order_date': order_date, 'title' : header, 'review_contents': review}

        # Append the dictionary to the data list
        data_list.append(data_dict)
        # print(data_list)
        # Convert the list of dictionaries into a DataFrame
#                 df = pd.DataFrame(data_list)
        time.sleep(5)

print("--------------------------------")
```
