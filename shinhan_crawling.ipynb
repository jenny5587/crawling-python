```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

# 크롬 드라이버 객체 생성
driver = webdriver.Chrome()

# 페이지 열기
url = "https://www.shinhan.com/hpe/index.jsp#050101010000"
driver.get(url)

##click to 대출
driver.find_element(By.XPATH, '/html/body/div[1]/div[2]/div/div/div[2]/div[1]/div[1]/div/div[2]/ul/li[3]/a').click()
t_list=[]

for k in range(10,11):
    try:
        print("페이지 수 : ",k)
        print("--------------------------------")
        driver.find_element(By.XPATH,f'//*[@id="pl_pageNav_page_{k}"]').send_keys(Keys.ENTER)
        time.sleep(5)
        
        try:
            
            for i in range(1,11):
                print("질문 수 : ",i)
                q_path=f'html/body/div[1]/div[2]/div/div/div[2]/div[1]/div[1]/div/div[3]/div[2]/div/table/tbody/tr[{i}]/td[2]/nobr/a'
                driver.find_element(By.XPATH, q_path).click()
                time.sleep(5)
                ##get text
                dt_elements = driver.find_elements(By.XPATH,"/html/body/div[1]/div[2]/div/div/div[2]/div[1]/div[2]/div/div[1]/dl/dt/span[2]")
                dd_elements = driver.find_elements(By.XPATH, "/html/body/div[1]/div[2]/div/div/div[2]/div[1]/div[2]/div/div[1]/dl/dd/span[2]")

                dt_list = []
                dd_list = []
                for i in range(len(dt_elements)):

                    dt = dt_elements[i].text.replace("\n"," ").strip()
                    dd = dd_elements[i].text.replace("\n"," ").strip()
                    dt_list.append(dt)
                    dd_list.append(dd)
                    dict_elements={'Instruction':dt_list[0],'Context':'','Response':dd_list[0],'Category':'open_qa'}
                    dict_elements['JSON']=f'{dict_elements}'
                t_list.append(dict_elements)
                driver.find_element(By.XPATH, '/html/body/div[1]/div[2]/div/div/div[2]/div[1]/div[2]/div/div[2]/a').click()
                time.sleep(5)
        except Exception as e:
            print('질문 없습니다!')
                 
    except Exception as e:
        print('페이지 없습니다!')
    print("--------------------------------")
```

```python
import openai
import pandas as pd
# open ai 플랫폼에서 key 할당받아 작업
openai.api_key = "your-key"

file_name = ['shinhan.csv']
```

```python
for file in file_name:
    print("file name is : ",file)
    new_df=pd.read_csv(file)[['Instruction','Response']]
    new_df.columns=['Instruction','Context']
    new_df['Category']='summarization'
    new_df['Response']=''
    new_df['JSON']=''

    for i in range(len(new_df)):
        content=new_df['Context'][i] + ' 앞에 문장을 100글자 이내로 요약해주세요.'
        chat_completion = openai.ChatCompletion.create(model="gpt-3.5-turbo", 
                                               messages=[{"role": "system", "content": "너는 금융 분야의 전문가입니다."},
                                                         {"role": "user", "content": content}])
        time.sleep(25)
        
        res_content=chat_completion.choices[0].message.content
        new_df['Response'][i]=res_content
        new_df['JSON'][i]=str(new_df.loc[i,['Instruction','Context','Response','Category']].to_dict())
        new_df=new_df[['Instruction','Context','Response','Category','JSON']]
        
    new_df.to_csv('summarization_'+file,header=True,index=False) 
    print("save to file : ",'summarization_'+file)
    time.sleep(3)
```
