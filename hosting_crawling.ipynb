### 연장 크롤링

```python
import undetected_chromedriver as uc
import time
import re
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import csv
import pandas as pd

# 크롬 열기
driver = uc.Chrome()
url = "https://help.hosting.kr/hc/ko/categories/360002280272"
driver.get(url)

q_list=[]

##click to 연장

driver.find_element(By.XPATH, '/html/body/main/div/div/div[2]/section[2]/h3/a').click()

# time.sleep(5)
##click to 질문
for i in range(1,7):
    print (i)
    driver.find_element(By.XPATH, f"/html/body/main/div[2]/div/section/ul/li[{i}]/a").click()

    ##질문 답 수집
    dt_elements = driver.find_elements(By.XPATH,"/html/body/main/div[2]/div/section/article/header/h1")
    dd_elements = driver.find_elements(By.XPATH, "/html/body/main/div[2]/div/section/article/section/div/div[1]")
    dt_list = []
    dd_list = []
    for i in range(len(dt_elements)):

        dt = dt_elements[i].text.replace("\n"," ").strip()
        dd = dd_elements[i].text.replace("\n"," ").strip()
        dt_list.append(dt)
        dd_list.append(dd)
        dict_elements={'Instruction':dt_list[0],'Context':'','Response':dd_list[0],'Category':'open_qa'}
        dict_elements['JSON']=f'{dict_elements}'
    q_list.append(dict_elements)
    ##back to domain 
    driver.find_element(By.XPATH, '/html/body/main/div[2]/div/section/nav/ol/li[3]/a').click()
    time.sleep(5)
```
